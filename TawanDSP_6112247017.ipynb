```python
from __future__ import print_function, division

%matplotlib inline

import thinkdsp
import thinkplot

import numpy as np

from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
from IPython.display import display
```

```python
#Lab01
sin_sig = thinkdsp.SinSignal(freq=7017, amp=17, offset=0)
sin_sig.plot()
thinkplot.config(xlabel='Time (s)')
```

```python
#Lab02
tsignal = thinkdsp.TriangleSignal(500)
duration = tsignal.period*500
tsegment = tsignal.make_wave(duration, framerate = 10000)
tsegment.plot()

twave = tsignal.make_wave(duration, framerate = 10000)
twave.apodize()
twave.make_audio()

tspectrum = twave.make_spectrum()
tspectrum.plot()

twave.apodize()
twave.make_audio()
```

```python
#Lab02
ssignal = thinkdsp.SquareSignal(100)
duration = ssignal.period*5
ssegment = ssignal.make_wave(duration, framerate = 10000)
ssegment.plot()
thinkplot.config(ylim = [-1.05, 1.05], legend = False)

swave = ssignal.make_wave(duration, framerate = 10000)
swave.apodize()
swave.make_audio()
```

```python
mixsignal = ssignal + tsignal
duration = mixsignal.period*20
msegment = mixsignal.make_wave(duration, framerate = 10000)
msegment.plot()

mwave = mixsignal.make_wave(duration, framerate = 10000)
mwave.apodize()
mwave.make_audio()
```
